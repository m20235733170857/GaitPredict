```python
%env KERAS_BACKEND=cntk
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout, Activation, LSTM, TimeDistributed 
from sklearn import preprocessing
import math
import socket
from sklearn.preprocessing import MinMaxScaler
```

建立TCP For Unity,用於send

```python
# 定义Unity服务器的IP地址和端口号
unity_server_address = ('localhost', 12345)  # 这里的IP地址和端口号应该与Unity服务器的设置相匹配

# 创建一个TCP/IP socket
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 连接到Unity服务器
client_socket.connect(unity_server_address)

# 发送信号到Unity服务器
def send_signal(signal):
    message = signal.encode()  # 将信号转换为字节串
    client_socket.sendall(message)  # 发送信号
```

建立TCP For IMU接收,用於receive

```python
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.bind(('localhost', 23456))
server.listen(1)  # 允許 1 個連線
print("伺服器已啟動，等待連線...")

conn, addr = server.accept()
print(f"已連接來自 {addr} 的客戶端")
```

```python
model_HS = load_model('model_test_total_0225.h5') #detect HS
model_predictL = load_model(r"C:\Users\m2023\Desktop\vision\predict_model\modelSeq2Seq LSTM with Positional Encoding and Multi-Head Attention For 1 Input For Leftt20241223.keras") #predict IMU for left
model_predictR = load_model(r"C:\Users\m2023\Desktop\vision\predict_model\modelSeq2Seq LSTM with Positional Encoding and Multi-Head Attention For 1 Input For Right20241223.keras") #predict IMU for right


data_5610=np.array([[1,2,3,4,5,6,7]])
data_5611=np.array([[1,2,3,4,5,6,7]])

R_HeelStrike_Predict_Time=np.array([[1]])
L_HeelStrike_Predict_Time=np.array([[1]])

# 初始化狀態機狀態
state = "IDLE"  # 初始狀態
step_counter = 0  # 用於記錄新增的步數
start_length = 0


cue_step = 5 #提前幾步cue
TRIAL_NAME = '\\20240516-2HS'
output_file = r"C:\Users\m2023\Desktop\experience_data"
```

函數

```python
#預測函數
def predict_data10step(imu_data_40step, model_predict):
    """
    處理數據並生成標準化的輸出。

    參數:
        imu_data (numpy.ndarray): 原始數據，包含至少 40 行。
        model_predict (object): 預測模型，需實現 predict 方法。

    返回:
        numpy.ndarray: 標準化後的數據，shape 為 (1, 50, 6)。
    """
    # 取過去 40 步的 Y 軸角加速度
    past_data = imu_data_40step[len(imu_data_40step) - 40:len(imu_data_40step), 5].reshape(-1, 1)  # shape: (40, 1)

    # 標準化
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaler.fit_transform(past_data)  # 訓練 scaler
    past_data_scaled = scaler.transform(past_data)  # 標準化
    past_data_scaled = past_data_scaled.reshape(1, 40, 1)  # shape: (1, 40, 1)

    # 初始化解碼器輸入
    decoder_input_test = np.zeros((1, 10, 1))
    decoder_input_test[:, 0, :] = 0

    # 預測未來 10 步
    predictions_scaled = []
    for step in range(10):
        prediction_step = model_predict.predict([past_data_scaled, decoder_input_test])  # 預測
        predictions_scaled.append(prediction_step[:, step, 0])  # 收集當前步的預測結果
        if step < 9:  # 更新解碼器輸入
            decoder_input_test[:, step + 1, 0] = prediction_step[:, step, 0]
    
    # 將標準化的預測結果轉換回原始數據範圍
    predictions_scaled = np.array(predictions_scaled).reshape(-1, 1)  # shape: (10, 1)
    predictions_orig = scaler.inverse_transform(predictions_scaled)  # 轉回原始值

    # 創建未來 10 步的數據矩陣
    future_data = np.zeros((10, 7))  # 初始化為全 0，shape: (10, 7)，包含第一column時間
    future_data[:, 6] = predictions_orig.flatten()  # 將原始值填入第 6 個 column，其餘 column 保持為 0

    # 合併過去 40 步與未來 10 步數據
    full_data = np.vstack([imu_data_40step[len(imu_data_40step) - 40:len(imu_data_40step), : ], future_data])  # shape: (50, 7) 包含時間column

    return full_data


#偵測HS函數
def detectHS(imu_data_50step, model_HS):
    processed_data = np.zeros_like(imu_data_50step)
    # 保留第 1 列和第 6 列數據
    processed_data[:, 0] = imu_data_50step[:, 0]  # 第 1 列 時間
    processed_data[:, 5] = imu_data_50step[:, 5]  # 第 6 列 Y軸角速度

    input=preprocessing.scale(processed_data[len(processed_data)-50:len(processed_data),1:7])
    input=input.reshape(1,50,6)
    hs_level=model_HS.predict(input)

    if hs_level[0,49,0]>0.015 and imu_data_50step[len(imu_data_50step)-1,5]*180/3.14<-10:
        signal = 1 
        print('predictHS')
        
        block_count+=1 

    return signal

#紀錄預測HS的真實時間
def save_data( L_HeelStrike_Predict_Time, R_HeelStrike_Predict_Time):
    index_min=min(   len(R_HeelStrike_Predict_Time),  len(L_HeelStrike_Predict_Time)   )
    index_0=np.array([["rightfoot"]])
    ee = np.vstack([index_0,R_HeelStrike_Predict_Time[1:index_min]])
    index_0=np.array([["leftfoot"]])
    ff = np.vstack([index_0,L_HeelStrike_Predict_Time[1:index_min]])
    gg = np.hstack([ee,ff])
    gg_csv = pd.DataFrame(gg)
    gg_csv.to_csv(output_file + TRIAL_NAME + 'predicthstime.csv',header=0,index=0)
    print(f"預測結果已成功保存到 {output_file}")
```

```python
data = conn.recv(1024)  # 接收資料
message = data.decode()

if message.startswith("data_5610:"):
    # 處理data_5610 右
    latest_data_5610 = message.replace("data_5610:", "")
    data_5610=np.vstack([data_5610,latest_data_5610])

elif message.startswith("data_5611:"):
    # 處理data_5611 左
    latest_data_5611 = message.replace("data_5611:", "")
    latest_data_5611=np.vstack([data_5610,latest_data_5611])

elif message == "ready to predict right leg": 
    predict_state = 1

elif message == "ready to predict left leg":
    predict_state = 0

elif message == "END":
    save_data(L_HeelStrike_Predict_Time, R_HeelStrike_Predict_Time)


if predict_state == 1 and len(data_5610)>40:
    
    if state == "IDLE":
        #開始預測右腳未來10步
        data_5610_50step = predict_data10step(data_5610, model_predictR)
        signal = detectHS(data_5610_50step, model_HS)

        # 如果偵測到未來第 10 步有踏步動作，進入監控狀態
        if signal == 1:
            state = "MONITOR"
            start_length = len(data_5610)  # 記錄當前資料的長度

    elif state == "MONITOR":
        # MONITOR 狀態：監控新增資料步數
        new_steps = len(data_5610) - start_length

        if new_steps == cue_step:
            # 當新增資料達到cue_step步，發送信號
            send_signal("0,1")
            
        if new_steps == 10:
            R_HeelStrike_Predict_Time = np.vstack([R_HeelStrike_Predict_Time, data_5610[len(data_5610)-1,0]]) #紀錄預測HS時真實的時間
            start_length, new_steps = 0  # 清除起始長度
            state = "IDLE"  # 重置狀態
            


elif predict_state == 0 and len(data_5611)>40:

    if state == "IDLE":
        #開始預測右腳未來10步
        data_5611_50step = predict_data10step(data_5611, model_predictL)
        signal = detectHS(data_5611_50step, model_HS)

        # 如果偵測到未來第 10 步有踏步動作，進入監控狀態
        if signal == 1:
            state = "MONITOR"
            start_length = len(data_5611)  # 記錄當前資料的長度

    elif state == "MONITOR":
        # MONITOR 狀態：監控新增資料步數
        new_steps = len(data_5611) - start_length

        if new_steps == cue_step:
            # 當新增資料達到cue_step步，發送信號
            send_signal("1,0")
            
        if new_steps == 10:
            L_HeelStrike_Predict_Time = np.vstack([L_HeelStrike_Predict_Time, data_5611[len(data_5611)-1,0]]) #紀錄預測HS時真實的時間
            start_length, new_steps = 0  # 清除起始長度
            state = "IDLE"  # 重置狀態
```
